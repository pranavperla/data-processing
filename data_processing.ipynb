Removes the punctuations from the text in the csv file

```python
import pandas as pd
import re

def remove_punctuation(text):

    #pattern for recognizing punctuation marks
    punctuation_pattern = r'[^\w\s]'

    #replacing punctuation marks with empty string
    text_without_punctuation = re.sub(punctuation_pattern, '', text)

    return text_without_punctuation

df = pd.read_csv('/content/output final.csv')

df['clean_text_User'] = df['User'].apply(remove_punctuation)
df['clean_text_odin'] = df['Odinschool'].apply(remove_punctuation)

df.to_csv('file1.csv', index=False)
df. drop('Odinschool',axis=1,inplace=True)
df. drop('User',axis=1,inplace=True)
df. drop('Unnamed: 0',axis=1,inplace=True)

df.head()
```

Remove the email ids from all the columns of the csvs file

```python
import pandas as pd
import re

def remove_email_ids(text):

    #identifying the email pattren
    email_pattern = r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\b"

    email_ids = re.findall(email_pattern, text)

    for email_id in email_ids:
        #replacing the email into empty string
        text = text.replace(email_id, "")

    return text

df = pd.read_csv('/content/output final.csv')

columns_to_process = ['User', 'Odinschool']

for column in columns_to_process:
    df[column] = df[column].astype(str)
    df[column] = df[column].apply(remove_email_ids)

df.to_csv('noemail.csv', index=False)
```

```python
import spacy
import pandas as pd

df = pd.read_csv('/content/without_missing3.csv')

# Load the English model - this contains a large corpus of data related to english language
nlp = spacy.load('en_core_web_sm')

for i, row in df.iterrows():
    doc = nlp(row['clean_text_User'])

    for ent in doc.ents:
        #identifying if the words in the doc is a persons name or not
        if ent.label_ == 'PERSON':
            row['clean_text_User'] = row['clean_text_User'].replace(ent.text, '')     #replacing names with empty string

df.to_csv('modified_file.csv', index=False)
```

Remove all the special charecters

```python

def remove_special_characters(text):

    if isinstance(text, str):

        #pattren to identify special characters
        special_character_pattern = r'[^a-zA-Z0-9\s]'

        text_without_special_characters = re.sub(special_character_pattern, '', text)

        return text_without_special_characters

    return text

df = pd.read_csv('/content/lower+punc.csv')

df = df.applymap(remove_special_characters)

df.to_csv('nopunc+lower+special.csv', index=False)
df.head()
```

As there is no significant use of the numeric data identified we have removed all the numeric data

```python
def remove_numbers(text):

    if isinstance(text, str):
        text_without_numbers = re.sub(r'\d+', '', text)
        return text_without_numbers
    return text

df = pd.read_csv('nopunc+lower+special.csv')

df['clean_text_User'] = df['clean_text_User'].apply(remove_numbers)
df['clean_text_odin'] = df['clean_text_odin'].apply(remove_numbers)

df.to_csv('nopunc+lower+special+numbers.csv', index=False)
df.head()
```

As stop words pose no real use to the model training they have been removed

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')

# Download the stop words corpus - this corpus of data has all the stop words(stop words like and,are,is etc..)
nltk.download('stopwords')

df = pd.read_csv('/content/nopunc+lower+special+numbers.csv')

def remove_stop_words(text):
    if isinstance(text, str):
        tokens = word_tokenize(text)
        stop_words = set(stopwords.words('english'))
        filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
        text_without_stop_words = ' '.join(filtered_tokens)
        return text_without_stop_words
    return text

columns_to_process = ['clean_text_User', 'clean_text_odin']

for column in columns_to_process:
    df[column] = df[column].apply(remove_stop_words)

df.to_csv('/content/nopunc+lower+special+numbers+stopwords.csv', index=False)

df.head()
```

Here, the data was tokenized and lemmatized

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('punkt')

nltk.download('wordnet')

df = pd.read_csv('/content/nopunc+lower+special+numbers.csv')

def tokenize(text):
    if isinstance(text, str):                   #converting paragraphs into words
        tokens = word_tokenize(text)
        return tokens
    return []


def apply_lemmatization(tokens):
    if isinstance(tokens, list):
        lemmatizer = WordNetLemmatizer()             # converting words into their base words
        lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
        return lemmatized_tokens
    return []


columns_to_process = ['clean_text_User', 'clean_text_odin']

for column in columns_to_process:
    df[column] = df[column].apply(tokenize)

    df[column] = df[column].apply(apply_lemmatization)


df.to_csv('/content/nopunc+lower+special+numbers_tokenized_lemmatized.csv', index=False)
df.head()
```
